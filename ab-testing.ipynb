### Step 1: Formulate Hypothesis

**Null Hypothesis:** The Conversion rate of old and new page is same.

**Alternate Hypothesis:** The Conversion rate of old and new page is different.

**Confidence level**: 95% 
**Alpha** = 1-0.95 = 0.05

If the probability of observing the p-value< alpha, then we reject null hypothesis. In simpler terms, we want to be 95% confident, that the results observed between the two conversion rates for the pages are different, before we reject null hypothesis.

We will use two-tailed test for this.


**Variables:**

Independent Variable -

1) Control group - Show old page

2) Treatment group - New page

Dependent Variable - Conversion rate (0 or 1) 

**Sample Size**: We will not test for the entire population. We will choose a sample. Hence the results are an estimate of true rates.
The number of people (or user sessions) we decide to capture in each group will have an effect on the precision of our estimated conversion rates: the larger the sample size, the more precise our estimates (i.e. the smaller our confidence intervals), the higher the chance to detect a difference in the two groups, if present. On the other hand, the larger our sample gets, the more expensive (and impractical) our study becomes.

**Power Analysis:**

**Power of the test (1-beta) **- This represents the probability of finding a statistical difference between the groups in our test when a difference is actually present. This is usually set at 0.8 as a convention.

**Alpha value (Alpha)** - The critical value we set earlier to 0.05.

**Effect size** - How big of a difference we expect there to be between the conversion rates.

Let us assume we are happy with a difference of 2%, we can use 13% and 15% to calculate the effect size we expect.
